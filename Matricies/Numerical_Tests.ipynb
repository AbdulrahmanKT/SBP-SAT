```python

n = 7 
k = 6
x7 , w7 = sb.lgl(n)
x6 , w6 = sb.lgl(k)
#G = np.array([sb.lagrange(n,x7[i]) for i in range(n+1)])
L7 = np.array([sb.lagrange(k,x7[i]) for i in range(n+1)]) # Using the 
P6 = sb.sbp_p(k)
u76 = np.linalg.inv(P7)@(L7.dot(P6@u6))

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x7,np.zeros_like(x7), "o",label="$x_7$")
ax.plot(x6,np.zeros_like(x6), "x",label="$x_6$")
ax.plot(x7,u76, label="P6 Projected to P7")
ax.plot(x7,u7,label="P7 Solution")
ax.plot(x6,u6,label="P6 Solution")
ax.legend()
ax.grid()
ax.set_title("Projecting P7 Jump onto a P6 Solution")
plt.show()
```
